```python
#Formatea la cadena de entrada en un solo string, quitando saltos de linea e informacion innecesaria
import pprint
secuence=''
for l in lines[1:]:
  secuence+=l[:-1]
pprint.pprint(secuence)
```

```python
def busqueda(subcadena,cadena):
  lista=[]                                                      #Definimos una lista para guardar el par de indices que representan la ubicacion de cada recurrencia de la subcadena(target) que queremos buscar dentro de la cadena (secuence)  
  posicion=0                                                    #Contador para llevar en cuenta en que posicion de la cadena vamos para poder definir la ubicacion de las recurrencias del target
  contador=0                                                    #Contador que servira para llevar en cuenta en numero de recurrencias del target dentro de la cadena 
  posString=0                                                   #Contador que servira para verificar que cada caracter dentro del target este efectivamente dentro de la cadena 
  if subcadena=="":                                             #Verficamos si la subcadena esta vacia, ya que no tiene mucho sentido buscar el vacio en una cadena
    return print("Subcadena vacia!") 
  for i in cadena:                                              #Recorremos cada caracter de la cadena(secuence) dada
    if i==subcadena[posString]:                                 #Verificamos si cada caracter de la cadena concuerda con los caracteres de la subcadena para verificar si hay una recurrencia
      posString+=1                                              #En caso tal de que un caracter de la subcadena se encuentre dentro de la cadena se sumara en 1 al contador correspondiente para poder evaluar en la siguiente iteracion el siguiente caracter de subcadena
    else:
      posString=0                                               #Si no concuerda volvemos el contador 0 para empezar a evaluar desde el primer caracter la subcadena 
      if subcadena[posString] == i:                             #Verificar si el caracter de finalizacion de cadena erronea coincide con el primero de la subcadena
        posString = 1                                           
    if posString==len(subcadena):                               #Si el contador que representa el indice de aciertos de caracteres de la subcadena dentro de la cadena es igual al numero de caracteres de la subcadena significa que hay una recurrencia dentro de la cadena
      lista.append([posicion-len(subcadena)+1,posicion])        #Agregamos a una lista la ubicacion de inicio y final de la subcadena dentro de la cadena 
      contador+=1                                               #Aumentamos en numero de recurrencias de la subcadena 
      posString=0                                               #Volvemos en 0 para empezar la comparacion otra vez desde el primer caracter de la subcadena
    posicion+=1                                                 #Aumentamos la posicion actual en la cadena
  return print(contador, lista)




target=input("Ingrese la subcadena a buscar: ")
busqueda(target, secuence)
#target=ACAGTAAGTGACAACAGATGTTT       Problema= TCTAGCACGTGCTGGTAA


```
